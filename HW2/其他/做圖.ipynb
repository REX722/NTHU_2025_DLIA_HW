```python
import matplotlib.pyplot as plt
import numpy as np

T_max = [50, 100, 150]
lr_values = [0.0001, 0.001, 0.01]
metrics = ['train', 'val', 'test']

# 数据（按 train/val/test 分组）
data = {
    'train': {
        0.0001: [69.3, 73.5, 76.7],
        0.001: [75.7, 84.7, 88.4],
        0.01: [83.6, 86.8, 87.3]
    },
    'val': {
        0.0001: [64.2, 70.4, 70.4],
        0.001: [56.8, 80.2, 70.4],
        0.01: [80.2, 80.2, 84.0]
    },
    'test': {
        0.0001: [64.5, 61.3, 61.3],
        0.001: [64.5, 74.2, 74.2],
        0.01: [67.7, 77.4, 71.0]
    }
}

# 绘制三个子图
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for i, metric in enumerate(metrics):
    ax = axes[i]
    for lr in lr_values:
        ax.plot(T_max, data[metric][lr], marker='o', label=f'LR={lr}')
    ax.set_xlabel('T_max')
    ax.set_ylabel(f'{metric} Accuracy (%)')
    ax.set_title(f'{metric.capitalize()} Accuracy')
    ax.legend()
plt.tight_layout()
plt.show()
```
